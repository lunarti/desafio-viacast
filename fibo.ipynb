```python
from timeit import default_timer as time
from functools import reduce
import matplotlib.pyplot as plt

def fibo1(n: int) -> int:

    fibo = lambda n: [listaux.append(sum(listaux)) or listaux.pop(0) 
                  for listaux in [[0,1]] for _ in range(n)]
    return fibo(n)[-1]

def fibo2(n: int) -> int:

    fibo = lambda y: reduce(lambda x, _: x + [x[-1] + x[-2]], range(y - 2), [0, 1])
    return fibo(n)[-1]

def fibo3(n: int) -> int:

    fibo_list = [0, 1]
    any(map(lambda _: fibo_list.append(sum(fibo_list[-2:])),
                                     range(2, n)))
    return fibo_list[-1]
```

```python
fig, ax = plt.subplots()
nAmostra=range(100)
fibo1plot=[]
fibo2plot=[]
fibo3plot=[]

for x in nAmostra:
    start = time()
    result=fibo1(450)
    end = time()
    t=end-start
    fibo1plot.append(t)

    start = time()

    start = time()
    result=fibo2(450)
    end = time()
    t=end-start
    fibo2plot.append(t)

    start = time()

    start = time()
    result=fibo3(450)
    end = time()
    t=end-start
    fibo3plot.append(t)

ax.scatter(nAmostra,fibo1plot,c="blue",label='fibo1')
ax.scatter(nAmostra,fibo2plot,c="green",label='fibo2')
ax.scatter(nAmostra,fibo3plot,c="orange",label='fibo3')
ax.legend()
plt.show()
```
